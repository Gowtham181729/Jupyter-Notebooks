```python
import numpy as np
import matplotlib.pyplot as plt
v = np.array([1,0])
w = np.array([1,2])
# w = -3 * v
print(w)

# Plot w
u=v+w;
origin = [0], [0]
plt.xlim(-2, 10)
plt.ylim(-2, 8)
name1="A+B"
# plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
mag_A = np.linalg.norm(v)
mag_B = np.linalg.norm(w)
mag_AB = np.linalg.norm(u)
mag_ab=v@w

print(mag_A, mag_B, mag_AB)
print(f"Since |A| * |B| >=|A.B| {mag_A}*{mag_B}>={mag_ab}")
print("Hence It satisifies Schawartz Inequlity")

plt.quiver(*origin, *v,scale=1,units='xy',label=f"|A|: {mag_A}")
plt.quiver(*origin, *w,scale=1,color='r',units='xy',label=f"|B|: {mag_B}")
plt.quiver(*origin, *u,scale=1,color='g',units='xy',label=f"|A+B|: {mag_AB}")
plt.text(5, 5, 'Triangular Inequalty', fontsize = 10, 
         bbox = dict(facecolor = 'red', alpha = 0.5))
plt.legend()
plt.grid()
plt.show()
```

```python
# Import required modules
import numpy as np
import matplotlib.pyplot as plt

# 1D arrays
x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

# Meshgrid
X,Y = np.meshgrid(x,y)

# Assign vector directions
Ex = (X + 1)/((X+1)**2 + Y**2) - (X - 1)/((X-1)**2 + Y**2)
Ey = Y/((X+1)**2 + Y**2) - Y/((X-1)**2 + Y**2)

# Depict illustration
plt.figure(figsize=(10, 10))
plt.streamplot(X,Y,Ex,Ey, density=1.4, linewidth=None, color='#A23BEC')
plt.plot(-1,0,'-or')
plt.plot(1,0,'-og')
plt.title('Electromagnetic Field')

# Show plot with grid
plt.grid()
plt.show()
```

```python
import numpy as np 

import pandas as pd 

import matplotlib.pyplot as plt 

# from pandas import DataFrame 

from sklearn import datasets 

from sklearn.mixture import GaussianMixture
```

```python
iris = datasets.load_iris()  

X = iris.data[:, :2]   

d = pd.DataFrame(X) 

plt.scatter(d[0], d[1])
```

```python
d
```

```python
gmm = GaussianMixture(n_components = 3)  

gmm.fit(d)  

labels = gmm.predict(d) 

d['labels']= labels 

d0 = d[d['labels']== 0] 

d1 = d[d['labels']== 1] 

d2 = d[d['labels']== 2] 

plt.scatter(d0[0], d0[1], c ='r') 

plt.scatter(d1[0], d1[1], c ='yellow') 

plt.scatter(d2[0], d2[1], c ='g')
```

```python
print("The mean:",gmm.means_)
print('\n')
print("The covariance matrix:",gmm.covariances_) 
print("The converged log-likelihood value:",gmm.lower_bound_) 
# print the number of iterations needed 
# for the log-likelihood value to converge 
print("The number of iterations needed for the log-likelihood value to converge:",gmm.n_iter_)
```
